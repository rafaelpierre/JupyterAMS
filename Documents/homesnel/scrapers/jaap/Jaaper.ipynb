```python
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
from nltk.tokenize import WordPunctTokenizer
from nltk.tokenize import RegexpTokenizer
```

```python
df_apartments_ams['stem'] = df_apartments_ams.description.str.findall('[A-Z|0-9|a-z]{2,}').str.join(' ').str.lower()
```

```python
df_apartments_ams['stem'] = df_apartments_ams.stem.apply(lambda x: [item for item in x.split(' ') if item not in stopwords.words('english')]).str.join(' ')

```

```python
import numpy as np

# Simple function to that returns a NaN if it is not fed a dict as an input.
def get_value(dict, string_to_get):
    '''
    takes input of dict, and tries to return the value of the string, if it fails
    it will return null value
    '''
    try:
        get_string = dict.get(string_to_get)
        return get_string
    except:
        return np.nan

df_apartments_ams['Price'] = [get_value(x,'price') for x in df_apartments_ams['property']]
df_apartments_ams['Area'] = [get_value(x,'area') for x in df_apartments_ams['property']]
```

```python
df_apartments_ams['Area'] = pd.to_numeric(df_apartments_ams.Area, errors='coerce')
```

```python
df_apartments_ams['location'] = [get_value(x,'latLng') for x in df_apartments_ams['property']]
```

```python
df_apartments_ams['Latitude'] = [get_value(x,'latitude') for x in df_apartments_ams['location']]
df_apartments_ams['Longitude'] = [get_value(x,'longitude') for x in df_apartments_ams['location']]
```
