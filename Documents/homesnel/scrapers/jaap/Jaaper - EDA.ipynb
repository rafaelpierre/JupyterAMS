```python
df['area'] = df.area.replace('Onbekend', '0')
df['lotSize'] = df.lotSize.replace('Onbekend', '0')

df['area'] = pd.to_numeric(df['area'])
df['lotSize'] = pd.to_numeric(df['lotSize'])
```

```python
import folium
```

```python
df = df[~pd.isnull(df.latLng)]
```

```python
df['lat'] = [d['latitude'] for d in df.latLng]
```

```python
df['lng'] = [d['longitude'] for d in df.latLng]
```

```python
df = df.dropna(axis=0, subset=['price', 'area'])
```

```python
df['price_per_meter'] = df.price / df.area
```

```python
df.price_per_meter.describe()
```

```python
m = folium.Map([52.3733752,4.8864105], zoom_start=12, tiles='Stamen Toner')

for i,row in df.iterrows():
    #print('s')
    red = int((row.price_per_meter / (df.price_per_meter.mean() + 2*df.price_per_meter.std())) * 255)
    green = 255 - red
    html = '#%02x%02x%02x' % (red, green, 0) #convert rgb to html
    
    house_string = '{} - {}m2'.format(row.price, row.area)
    
    marker = folium.CircleMarker(location=[row.lat,row.lng], radius=(row.area/25), color=html, fill=html, fill_opacity='0.3', popup=house_string)
    m.add_child(marker)

marker = folium.Marker(location=[52.3670621,4.8680757])

m.add_child(marker)
        
m
```
